```python
# Step1
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Step2
fig, ax = plt.subplots(2, 3, tight_layout=True)

# Step3

# 折れ線グラフ
y0 = [1, 2, -5, 2]
ax[0, 0].plot(y0)

# sin関数の設定
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)
ax[0, 1].plot(x1, y1, linewidth=2.0)
ax[0, 1].set(xlim=(0, 10), xticks=np.arange(0, 11),
             ylim=(-1, 5), yticks=np.arange(-1, 6))

# ヒストグラム
x2 = np.random.normal(15, 5, 2000)
ax[0, 2].hist(x2, bins=20)

# 散布図
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))
ax[1, 0].scatter(x3, y3, marker='o')

# Step4
plt.show()
```
